The Collatz Conjecture (1937): Start with any positive integer n, divide by two if n is even, or multiply by three and add one if n is odd, repeat. The conjecture states that this sequence will eventually reach 1. Almost a century later, the conjecture is still unsolved: no mathematical proof and no counter-example! Studying it nevertheless has lead into several directions intersectsing number theory, combinatorics, mathematical analysis, probability and statistics, fractal geometry, and computational complexity. Paul Erdos famously said "Mathematics may not be ready for such problems". 
Below is a simple python script to calcualte and visualize the Collatz sequence. Do you see a common pattern there?! 
Read more about the conjecture here (https://en.wikipedia.org/wiki/Collatz_conjecture).  

```python
import math
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def collatz_sequence(n):
    if not isinstance(n, int) or n <= 1:
        print("Invalid input, please provide an integer greater than 1!")
    else: 
        s = [n]
        while(n > 1):
            if (n%2) ==0:
                n = int(n/2) 
            else:
                n = int(3*n + 1)
            s.append(n)
        return s
```

```python
collatz_sequence(10)
```

```python
collatz_sequence(17)
```

```python
collatz_sequence(15)
```

```python
#this may take a while to run
coll_seq_df = pd.DataFrame()
for n in range(2,50000,10):
    df = pd.DataFrame(data={'n':n,'Collatz_Sequence_Length':collatz_sequence(n)})
    coll_seq_df = coll_seq_df.append(df,ignore_index=True)
```

```python
plt.figure(figsize=(15,8))
sns.set_style('darkgrid')
sns.scatterplot(data=coll_seq_df,x='n',y='Collatz_Sequence_Length', hue='n',palette=plt.cm.brg);
plt.title("length of collatz sequence for integer n");
```

```python
n1=random.randint(2,1e3)
n2=random.randint(2,1e4)
n3=random.randint(2,1e5)

plt.figure(figsize=(15,8))
sns.set_style('darkgrid')
plt.plot(collatz_sequence(n1),color='black', marker='s', linestyle='dashed',linewidth=1, markersize=2, label= 'n = %i' %n1);
plt.plot(collatz_sequence(n2),color='red',  linestyle='dashed',linewidth=1, markersize=3, label= 'n = %i' %n2);
plt.plot(collatz_sequence(n3),color='green', marker='s', linestyle='dashed',linewidth=1, markersize=2, label= 'n = %i' %n3);

plt.legend(loc="upper right");
plt.xlabel("Collatz Sequence Length");
plt.ylabel("Collatz Sequence Value");
```
